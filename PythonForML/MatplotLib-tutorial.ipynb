```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
## simple exaples

x = np.arange(0,10)
y = np.arange(11, 21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
## plotting using matplotlib

## scatter plot

plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graphs in 2D')
```

```python
## plt plot
y = x*x

plt.plot(x,y, 'ro', linestyle='dashed', linewidth=2, markersize=6)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D diagram')
```

```python
## creating subplots

## subplot(rows, columns, position)

plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'b')
plt.subplot(2,2,4)
plt.plot(x,y,'g')
```

```python
x = np.arange(1,11)
y = 3*x + 5

plt.plot(x,y)
```

```python
# compute x and y co ordinate for points on a sine curve

x = np.arange(0, 8 * np.pi, 0.1)
y = np.sin(x)

plt.plot(x,y)
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin, 'r')
plt.title("sine wave")

plt.subplot(2,1,2)
plt.plot(x,y_cos, 'g')
plt.title("cos wave")
```

```python
## Bar plots

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2, color='g')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar graph")
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a, bins=5)
plt.title("histogram")
```

# Box plot using matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]

# rectangular box plot
plt.boxplot(data, vert=True, patch_artist=False)
```

# PIE chart

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215,130,245,210]
colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1,0,0,0)

plt.pie(sizes, explode=explode, labels = labels, colors=colors, autopct='%1.1f%%', shadow=True)
```
